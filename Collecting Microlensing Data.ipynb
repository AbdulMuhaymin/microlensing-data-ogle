```python
import gzip
import urllib.request
import io
import tarfile
```

```python
import pandas as pd
```

```python
df = pd.read_csv('ogle_planets.csv', skiprows=260, header=1)
```

```python
names = list(df['pl_name'])
```

```python
name = []
for i in range(len(names)):
    ii = names[i].find('L', 16)
    name.append(names[i][:ii])
    name = sorted(name)
```

```python
import requests
```

```python
for i in name:
    year = i[5:9]
    num = i[14:]
    if int(year) < 2010:
        version = 3
    else:
        version = 4
    
    url = "https://www.astrouw.edu.pl/ogle/ogle" + str(version) + "/ews/" + year + "/blg-" + str(num)+ ".tar.gz"
    
    print(url)
    
    response = urllib.request.urlopen(url)
    compressed_file = io.BytesIO(response.read())
    decompressed_file = gzip.GzipFile(fileobj=compressed_file)
    
    out = "gzips/" + i + ".gzip"
            
    with open(out, 'wb') as outfile:
        outfile.write(decompressed_file.read())
    
    file = tarfile.open(out)

    file.extractall("extracted/")

    file.close()
```
